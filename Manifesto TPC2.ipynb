### Resumo

* O programa começa por pedir ao jogador para pensar num número entre 0 e 100 e define 4 variáveis: 3 inteiros e 1 boleano.
* 2 das variáveis são inicializadas no mesmo valor (50), de forma a manter uma para as sucessivas divisões e outra à qual as divisões serão somadas, que no fim será o número escolhido pelo jogador. O restante inteiro serve de contagem de tentativas e o boleano como fator de paragem no ciclo while.
* Dentro do ciclo while o computador vai somar ou subtrair o valor dividido ao valor somado, desta forma aproximando-se cada vez mais do valor a adivinhar.
* Foram ainda introduzidas algumas correções relativas às tentativas 5 e 6, fruto do resultado divisão inteira, que ao arredondar introduzia um erro; e à possibilidade do jogador tentar enganar o computador.
